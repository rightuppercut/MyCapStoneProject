```python
import pandas as pd
import numpy as np
import wikipedia as wp
import requests
import os
import urllib.request
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
import folium
print("Hello Capstone Project Course!")
```

```python
url = "https://en.wikipedia.org/wiki/Demographics_of_Toronto_neighbourhoods"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
all_tables=soup.find_all("table")

print(soup.title.string)
right_table=soup.find('table', class_='wikitable sortable')
#print(right_table)

A = []
B = []
C = []
D = []
E = []
F = []
G = []
H = []
I = []
J = []
K = []
L = []

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==13:
        A.append(cells[0].find(text=True).rstrip("\n"))
        B.append(cells[1].find(text=True).rstrip("\n"))
        C.append(cells[2].find(text=True).rstrip("\n"))
        D.append(cells[3].find(text=True).rstrip("\n"))
        E.append(cells[4].find(text=True).rstrip("\n"))
        F.append(cells[5].find(text=True).rstrip("\n"))
        G.append(cells[6].find(text=True).rstrip("\n"))
        H.append(cells[7].find(text=True).rstrip("\n"))
        I.append(cells[8].find(text=True).rstrip("\n"))
        J.append(cells[9].find(text=True).rstrip("\n"))
        K.append(cells[10].find(text=True).rstrip("\n"))
        L.append(cells[11].find(text=True).rstrip("\n"))
        
Wiki_DF = pd.DataFrame(A,columns=['Name'])
Wiki_DF['FM']=B
Wiki_DF['Census Tracts']=C
Wiki_DF['Population']=D
Wiki_DF['Land Area']=E
Wiki_DF['Density']=F
Wiki_DF['Change in Population']=G
Wiki_DF['Average Income']=H
Wiki_DF['Transit']=I
Wiki_DF['Renters']=J
Wiki_DF['Snd Most Common Language']=K
Wiki_DF['SMCL %']=L

Wiki_DF.head()
Wiki_DF.to_csv(os.getcwd() + "\\MyTorontoData.csv")
```

```python
# create map of New York using latitude and longitude values
FMDict = {"S":"Scarborough", "OCoT":"Old City of Toronto", "NY":"North York", "E":"Etobicoke", "Y":"York", "EY":"East York"}
column_names = ['FM', 'Neighborhood', 'Latitude', 'Longitude'] 
neighborhoods = pd.DataFrame(columns=column_names)

MY_DF_N = Wiki_DF.loc[Wiki_DF['FM'] == 'S']

for i,(index,row) in enumerate(MY_DF_N.iterrows()):
    #if i == 0: continue # skip first row
    Name = row['Name']
    FM = FMDict[row['FM']]
    Place = Name + ", Toronto"
    geolocator = Nominatim(user_agent='MyCapStone')
    location = geolocator.geocode(Place)
    LAT = location.latitude
    LON = location.longitude
    neighborhoods = neighborhoods.append({'FM': FM,
                                              'Neighborhood': Name,
                                              'Latitude': str(LAT),
                                              'Longitude': str(LON)}, ignore_index=True)
geolocator = Nominatim(user_agent="MyCapStone")
location = geolocator.geocode("Toronto")
latitude = location.latitude
longitude = location.longitude

map_Toronto_Scarborough = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['FM'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto_Scarborough)  
    
map_Toronto_Scarborough
```
